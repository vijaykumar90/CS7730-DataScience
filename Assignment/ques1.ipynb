```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

```

```python
data = 'diabetes_01_health_indicators.csv'
```

```python
df = pd.read_csv(data)
print("Shape of DataFrame: ", df.shape)
df.isnull().sum()
```

```python
D = df.drop("Diabetes_binary", axis=1)
L = df["Diabetes_binary"]
D.shape
```

```python
D.keys()
```

```python
L.value_counts()
```

```python
import pandas as pd
import seaborn as sns

# Simulate your Series
# L = pd.Series([0.0, 1.0, 0.0, 1.0, 0.0])  # uncomment for test
split = L.value_counts()

# Create DataFrame with a hue column
bardf = pd.DataFrame({
    'Diabetes_binary Labels': ['0.0', '1.0'],
    'Frequency': [split[0.0], split[1.0]],
})

# Add hue column for color mapping (same as label, but string)
bardf['Label Class'] = bardf['Diabetes_binary Labels']  # This will act as `hue`

# Create plot using hue
sns.barplot(
    data=bardf,
    x="Diabetes_binary Labels",
    y="Frequency",
    hue="Label Class",
    palette={'0.0': 'red', '1.0': 'white'}
).set_title('Diabetes_binary Class Distribution')
```
